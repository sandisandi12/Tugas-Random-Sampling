```python
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv("MovieData-2018-2020.csv")
data
```

```python
data['genre'].value_counts()
```

```python
#hilangkan genre yang hanya berisi 1 judul
filtered_data = data.groupby('genre').filter(lambda x: len(x) > 1)
filtered_data['genre'].value_counts()
```

Pakailah hanya 30 data teratas dari hasil sampling

```python
#stratified sample by genre
_, stratified_sample = train_test_split(
    filtered_data,
    test_size = 0.3,
    stratify = filtered_data ['genre'],
    random_state = 49
)
sample = stratified_sample.head(30)
data_budget = sample['budget']

print("Stratified Sampling (30 data teratas):\n", sample)
```

```python
stratified_sample.describe()
```

Hitunglah Estimasi (Statistics - Estimating Population Means) Budget Film menggunakan data sample tadi

```python
import scipy.stats as stats
import math

x_bar = data_budget.mean()
s = data_budget.std(ddof=1)  #menggunakan ddof=1 karena ini sebuah data sample
n = len(data_budget)
confidence_level = 0.95

# Calculate alpha, degress of freedom (df), the critical t-value,
# And the margin of error
alpha = (1 - confidence_level)
df = n - 1
standard_error = s / math.sqrt(n)
critical_t = stats.t.ppf(1 - alpha / 2, df)
margin_of_error = critical_t * standard_error

# Calculate the lower and upper bound of the confidence interval
lower_bound = x_bar - margin_of_error
upper_bound = x_bar + margin_of_error

# Print the results
print("Critical t-value : {:.2f}".format(critical_t))
print("Margin of Error : {:.2f}".format(margin_of_error))
print("Confidence Interval : ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
print("The {:.1%} confidence interval for the population mean is:".format(confidence_level))
print("between {:.2f} and {:.2f}".format(lower_bound, upper_bound))

```
